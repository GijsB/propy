```python
import propy as pp
import numpy as np
import matplotlib.pyplot as plt
```

```python
pds = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4]
j = np.linspace(0, 1.6, 1000)

plt.figure()
for i, pd in enumerate(pds):
    prop = pp.WageningenBPropeller(
        diameter= 1,
        blades=2,
        pd_ratio=pd,
        area_ratio=0.3
    )
    
    plt.plot(j, prop.kt(j), color=f'C{i}', linestyle='-')
    plt.plot(j, 10*prop.kq(j), color=f'C{i}', linestyle='--')
    plt.plot(j, prop.eta(j), color=f'C{i}', linestyle=':')
    

plt.grid()
plt.ylim(0, 1.6)
```
